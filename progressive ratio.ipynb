```python
def plot_pressing(data, ani_num, ax = None):

    if not ax:
        fig, ax = plt.subplots(figsize = (8, 3))
    

    d1_press = data.loc[data.Event == ' door_1 lever pressed', 'Time']
    d2_press = data.loc[data.Event == ' door_2 lever pressed', 'Time']

    new_round = data.loc[data.Event == ' Starting new round', 'Time']
    d1_reward = data.loc[data.Event == ' door_1 lever pressed productive', 'Time']
    d2_reward = data.loc[data.Event == ' door_2 lever pressed productive', 'Time']
    

    ax.vlines(d1_press, ymin = 0, ymax = 1, color = '#ff00ff', lw = 0.2)
    ax.vlines(d2_press, ymin = -1, ymax = 0, color = '#0000FF', lw = 0.2)

    ax.vlines(d1_reward, ymin = 0, ymax = 1, color = '#ff00ff', lw = 0.2)
    ax.vlines(d2_reward, ymin = -1, ymax = 0, color = '#0000FF', lw = 0.2)
    #ax.vlines(new_round, ymin = 0.8, ymax = 1.2, color = 'gray')
    #ax.vlines(new_round, ymin = -0.8, ymax = -1.2, color = 'gray')
    ax.vlines(new_round, ymin = -0.1, ymax = 0.1, color = 'black')
    ax.scatter(d1_reward, [0]*len(d1_reward),  marker = 'o', color = '#ff00ff')
    ax.scatter(d2_reward, [0]*len(d2_reward),  marker = 'o', color = '#0000ff')

    ax.set_xlabel('time')
    ax.set_yticks(())
    ax.set_title(ani_num)
```

```python
names = af.assemble_names('/Users/davidprotter/Documents/Donaldson Lab/Sex_diff_c1/csv')
pr_files = [f for f in names if 'prog' in f if '19' in f if not 'summary' in f if not 'round' in f]
```

```python
fig, axs = plt.subplots(nrows = 4, ncols = 3, figsize = (10, 15))

for f, ax in zip(pr_files, axs.ravel()):
    _,_,df = ana.prep_for_analysis(f)
    vole = af.get_header(f)['vole']

    num = f'{vole} (F)' if vole in females else f'{vole} (M)'

    plot_pressing(df, num, ax)
fig.tight_layout()
```
